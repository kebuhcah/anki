```python
import pandas as pd
from pandas import DataFrame
```

```python
ChuNom = pd.read_csv('standard-list.txt', delimiter='\t').sort_values('Freq.')
ChuNom = ChuNom[(~ChuNom.Unicode.isnull())&(~ChuNom.English.isnull())].reset_index(drop=True)
ChuNom
```

```python
ChuNomExport = ChuNom[['Text','QN', 'English', 'Unicode']]
ChuNomExport.columns = ['text', 'qn', 'english', 'unicode']
#ChuNomExport.to_csv('anki.tsv', sep='\t', index=False)
ChuNomExport
```

```python
with open('anki_2.tsv', 'w') as f:
    for _, r in ChuNomExport.iterrows():
        s = "%s\t%s<br>%s\n" % (r.text, r.qn, r.english)
        print s,
        f.write(s)
```

```python
with open('get_glyphs.sh', 'w') as f:
    for uc in ChuNomExport.unicode:
        codepoint = uc[2:]
        cmd = "curl https://www.chunom.org/media/generated/%s-32.png --output glyphs/%s.png" % (codepoint, codepoint)
        print cmd
        f.write(cmd)
        f.write('\n')
```

```python
with open('anki_w_imgs.tsv', 'w') as f:
    for _, r in ChuNomExport.iterrows():
        s = "<img src=\"%s.png\"><br>%s\t%s<br>%s\n" % (r.unicode[2:], r.text, r.qn, r.english)
        print s,
        f.write(s)
```
